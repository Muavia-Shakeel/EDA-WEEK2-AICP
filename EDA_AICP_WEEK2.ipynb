Q.1:  
    Writer pandas series code to get following output without using dictionary:
    
    
    a   1

    x   4

    c   9

    2   6

    e   7

    dtype: int64


```python
import pandas as pd

# Data list
data = [1, 4, 9, 6, 7]
index_labels = ['a', 'x', 'c', '2', 'e']

# Creating a pandas Series from the list
series_without_dict = pd.Series(data, index=index_labels)

print("pandas Series without using a dictionary\n", series_without_dict)
```
